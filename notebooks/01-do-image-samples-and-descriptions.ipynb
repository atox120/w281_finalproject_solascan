**Closed Crack**: Black narrow line in the image

**Resistive Crack**: Black closed narrow line with darkened regions extending from the crack

```python
img_load.load_n(3, defect_classes='NearSolderPad')
def_view.view_defects(sample_df=None, group_by='defect_class', annotation_type='segmentations')
```

**Isolated Crack**: Regions that are clearly individual sections of bulk material that have been split from the main cell by cracks

```python
img_load.load_n(3, defect_classes='Isolated')
def_view.view_defects(sample_df=None, group_by='defect_class', annotation_type='segmentations')
```

**Front Grid Interupption**: darkened rect- angles appearing along finger lines

```python
img_load.load_n(3, defect_classes='FrontGridInterruption')
def_view.view_defects(sample_df=None, group_by='defect_class', annotation_type='segmentations')
```

**Disconnected Interconnections**: long dark rectangles that extend along the entire length of the busbar

```python
img_load.load_n(3, defect_classes='Disconnected')
def_view.view_defects(sample_df=None, group_by='defect_class', annotation_type='segmentations')
```

**HighlyResistive Interconnections**: When the entire busbar is bright, the bright EL response extends a short distance perpendicular to the busbar, but extends farther when it is only part of the busbar. Several of the highly resistive interconnects in our dataset do not affect the entire length of the busbar. This gives shorter (less distance across the busbar) but wider (longer perpendicular distance) bright regions.

```python
img_load.load_n(3, defect_classes='HighlyResistive')
def_view.view_defects(sample_df=None, group_by='defect_class', annotation_type='segmentations')
```

**Resistive Interconnections**:

```python
img_load.load_n(3, defect_classes='Resistive')
def_view.view_defects(sample_df=None, group_by='defect_class', annotation_type='segmentations')
```

**Rear Contact Belt Marks**: due to nonuniform firing of the rear Al paste. The darker regions observed are normally the location where the moving belt of the fast firing furnace contacts the paste.

```python
img_load.load_n(3, defect_classes='BeltMarks')
def_view.view_defects(sample_df=None, group_by='defect_class', annotation_type='segmentations')
```

**Contact Corrosion**: has a characteristic darkening pattern starting at the busbar and extending perpendicularly

```python
img_load.load_n(3, defect_classes='Corrosion')
def_view.view_defects(sample_df=None, group_by='defect_class', annotation_type='segmentations')
```
